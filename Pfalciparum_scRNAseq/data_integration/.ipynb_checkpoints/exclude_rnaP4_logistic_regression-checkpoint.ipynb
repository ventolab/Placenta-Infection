```python
adataRef.obs['clusters_IDC_1'] = adataRef.obs['Stage']
#adataRef.obs['clusters_IDC_2'] = adataRef.obs['pbbulk_hoo_etal']
```

### Filter common genes,  using either HVG overlap or non-HVG overlap

```python
common_genes = list(set(adataRef.var.index.values).intersection(adataTest.var.index.values))
print('common genes: ', len(common_genes))
adataTest = adataTest[:,common_genes]
adataRef = adataRef[:,common_genes]
print(adataRef.X.shape)
print(adataTest.X.shape)
```

```python
# list of genes shared data
# doing sorted on the overall list is VERY important for consistent ordering of the genes
common_genes = sorted(list(set(adataRef.var_names) & set(adataTest.var_names)))
print(len(common_genes),"genes are shared between annotated train and our joint data \n",
      len(set(adataRef.var_names)), 'genes in training data \n',
      len((set(adataTest.var_names))), 'genes in test joint data \n')
```

```python
#Subseting the datasets
adataTest_copy = adataTest[:,common_genes]
adataTest_copy.var_names
```

```python
#Subseting the datasets
adataRef_copy = adataRef[:,common_genes]
adataRef_copy.var_names
```

```python
#normlog and scaling Training and Test sets
sc.pp.normalize_total(adataTest, target_sum=1e4)
sc.pp.log1p(adataTest)
sc.pp.scale(adataTest, max_value=10)

sc.pp.normalize_total(adataRef, target_sum=1e4)
sc.pp.log1p(adataRef)
sc.pp.scale(adataRef, max_value=10)
```

### Build a regression classifier from MCA ref dataset

```python
logisticRegr = LogisticRegression(max_iter = 4000, n_jobs = -1, random_state = 0, C=0.2)
logisticRegr.fit(adataRef.X, [str(i) for i in adataRef.obs['clusters_IDC_1'].values])
```

### Classify predictions in test samples

```python
predictions = logisticRegr.predict(adataTest.X)
adataTest.obs['predictions'] = predictions
probabilities = logisticRegr.predict_proba(adataTest.X)
probs_adataTest = []
for i,clus in enumerate(adataTest.obs['predictions']):
    probs_adataTest.append(probabilities[i,logisticRegr.classes_==clus][0])
```

```python
adataTest.obs['probabilities'] = probs_adataTest
#pf_adata.obsm["X_umap"] = pf_adataP.obsm["X_umap_noDiff"]
adataTest.obsm["X_umap"] =adataTest.obsm["X_umap"]
sc.pl.umap(adataTest, color=['probabilities', 'predictions' ])
```

```python
adataTest.obs['idc_new'] = adataTest.obs['predictions'].tolist()
adataTest.obs.at[adataTest.obs['idc_new'] == 'late ring','idc_new'] = '2_late_ring'
adataTest.obs.at[adataTest.obs['idc_new'] == 'early ring', 'idc_new'] = '1_early_ring'
adataTest.obs.at[adataTest.obs['idc_new'] == 'early trophozoite', 'idc_new'] = '3_early_troph'
adataTest.obs.at[adataTest.obs['idc_new'] == 'late trophozoite', 'idc_new'] = '4_late_troph'
adataTest.obs.at[adataTest.obs['idc_new'] == 'early schizont', 'idc_new'] = '5_early_schiz'
adataTest.obs.at[adataTest.obs['idc_new'] == 'late schizont', 'idc_new'] = '6_late_schiz'
set(adataTest.obs['idc_new'])
```
