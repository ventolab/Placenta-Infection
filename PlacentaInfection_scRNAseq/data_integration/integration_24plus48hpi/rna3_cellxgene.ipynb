```python
#Selecting useful metadata
adata.obs= adata.obs[['n_counts','n_genes','percent_mito','cell_type','hpi','stage','phase','donor_id','MFgenotype','infection',
           'stage_perInfection','Tg_infected','Dev_Stage']]

#renaming this column
#adata.obs= adata.obs.rename(columns={"MFgenotype": "origin_M_F"})

#renaming this column
#adata.obs= adata.obs.rename(columns={"identity": "CellType"})
```

```python
#removing unnecessary metadata
del(adata.uns)
del(adata.obsp)
```

```python
#Log-normalising the counts
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
```

```python
adata.X = scipy.sparse.csr_matrix(adata.X)
```
