```python
# Define the log2FC threshold for differential expression
log2FC_threshold = 0.2

# Filter genes based on the log2FC threshold and significance level (if available)
filtered_df = df1[(abs(df1['avg_log2FC']) >= 0)]

# Volcano plot
plt.figure(figsize=(6, 6))
plt.scatter(filtered_df['avg_log2FC'], -1 * np.log10(filtered_df['p_val_adj']), color='green', alpha=0.5, s=15)

# Highlight differentially expressed genes
differentially_expressed_genes = filtered_df[(filtered_df['p_val_adj'] < 0.01) & (abs(filtered_df['avg_log2FC']) > log2FC_threshold)]
plt.scatter(differentially_expressed_genes['avg_log2FC'], -1 * np.log10(differentially_expressed_genes['p_val_adj']),
            color='red', label='Differentially Expressed', alpha=0.7, s=15)


# Highlight non-differentially expressed genes
differentially_expressed_genes = filtered_df[(filtered_df['p_val_adj'] > 0.01) ]
plt.scatter(differentially_expressed_genes['avg_log2FC'], -1 * np.log10(differentially_expressed_genes['p_val_adj']),
            color='blue', label='Differentially Expressed', alpha=0.7, s=15)


# Customize the plot
plt.axvline(x=0.2, color='green', linestyle='--', label=f'Log2FC Threshold ({log2FC_threshold})')
plt.axvline(x=-0.2, color='green', linestyle='--')
plt.axhline(y=2, color='black', linestyle='--')


plt.xlabel('Log2FC')
plt.ylabel('-log10(p_val_adj)')
plt.title('Volcano Plot')
plt.legend()
plt.grid(True)

# Save or display the plot
#plt.savefig('volcano_plot.png', dpi=300)  # Save the plot as a PNG image
plt.show()
```

```python
counts = {}
for deg_sample in set(temp.obs['DEG_variable']):
    
    deg_sample
    
    group_samples = temp.obs[temp.obs['DEG_variable'] == deg_sample].index
    
    counts[deg_sample] = temp[group_samples].X.sum(axis = 0).tolist()
```

```python
%%R

design <- model.matrix(~ Group, y$samples)
y <- estimateDisp(y, design)
summary(y$trended.dispersion)
plotBCV(y)
```

```python
%%R

fit <- glmQLFit(y, design, robust = TRUE, sample.weights = TRUE)
plotQLDisp(fit)
```

```python
%%R -o tt

res <- glmQLFTest(fit, coef = 'GroupUI_Pf_24h',)

tt = res$table
tt$bonferroni = p.adjust(tt$PValue, "bonferroni")

hist(tt$PValue)
```
