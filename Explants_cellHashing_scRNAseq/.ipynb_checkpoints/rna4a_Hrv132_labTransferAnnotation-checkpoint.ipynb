# Notebook to explore cell type predictions product of the logistic regression from the MFI reference - Donor Hrv132

```python
import numpy as np
import pandas as pd
import scanpy as sc
import os
import sys
import anndata
import scvi
```

```python
def MovePlots(plotpattern, subplotdir):
    os.system('mkdir -p '+str(sc.settings.figdir)+'/'+subplotdir)
    os.system('mv '+str(sc.settings.figdir)+'/*'+plotpattern+'** '+str(sc.settings.figdir)+'/'+subplotdir)

sc.settings.verbosity = 3  # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.settings.figdir = './figures/annotation/'
sc.logging.print_versions()
sc.settings.set_figure_params(dpi=80)  # low dpi (dots per inch) yields small inline figures

sys.executable
```

## Read donor Hrv132 which already had predicitions labeled 

```python
adata= sc.read('rna3_counts_HTO_donorHrv132_logRegression_adata.h5ad')
```

```python
adata
```

```python
# use leiden imputed values for plotting
sc.pl.umap(adata, color=[ 'leiden','Classification','probabilities','celltype_predictions'])
```

```python
sc.pl.umap(adata, color=['doublet_scores','fraction_HTO3', 'fraction_HTO4', 'fraction_HTO5'] , color_map='OrRd')
```

## Cluster pre-labeling

The marker genes already reported for different cell types were intensively analysed, together with the clusters' markers, and the label trasfering.


```python
adata
```

## Re-run leiden clustering

```python
sc.pp.neighbors(adata, n_neighbors=15)
sc.tl.umap(adata)
```

```python
#clustering
sc.tl.leiden(adata, key_added="leiden", resolution=0.8)
```

```python
sc.pl.umap(adata, color=['leiden','Classification','probabilities','celltype_predictions','doublet_scores','n_genes', 'n_counts'],
           ncols=2, wspace=0.3)
```

## Broad annotation of cell clusters

```python
#Copying the value of the leiden cluster. The new column will help to change it's values for
# the cluster identity
adata.obs['identity']= adata.obs['leiden']
```

```python
adata.obs['identity']
```

```python
#ranking genes check top genes
sc.tl.rank_genes_groups(adata, groupby='leiden', method='logreg')
```

```python
sc.pl.rank_genes_groups(adata)
```

```python
adata.obs['identity'] = (adata.obs["leiden"].map(lambda x: 
                            {"0": "HB", "1": "HB", "2": "HB", "3": "Fibroblast","4": "Trophoblast", "5": "HB", 
                             "6": "lowQC", "7": "HB",  "8": "lowQC", "9": "Fibroblast",  "10": "PAMM1", "11": "HB", 
                              "12": "Trophoblast", "13": "Endo_f", "14": "Fibroblast",  "15": "scrublet_doublet",
                              "16": "scrublet_doublet", "17": "Erythrocyte", "18": "dS"}
                            .get(x, x)).astype("category"))
```

```python
sc.pl.umap(adata, color=['identity', ], legend_loc='on data',legend_fontsize='xx-small')
```

## Removal of scrublet doublets, doublet cluster, and Erythrocytes etc.

```python
adata = adata[adata.obs['identity'] != 'Erythrocyte']
adata = adata[adata.obs['identity'] != 'scrublet_doublet']
adata = adata[adata.obs['identity'] != 'dS']
adata = adata[adata.obs['identity'] != 'lowQC']
```

```python
sc.pl.umap(adata, color=['identity', ], legend_loc='on data',legend_fontsize='xx-small')
```

## Re-run leiden clustering

```python
sc.pp.neighbors(adata, n_neighbors=20)
sc.tl.umap(adata)
```

```python
#clustering
sc.tl.leiden(adata, key_added="leiden", resolution=1)
```

```python
sc.pl.umap(adata, color=["leiden", 'n_genes', 'Classification', 'identity', ],
           ncols=2, wspace=0.3)
```

```python
adata
```

```python
#Normalizing the data
sc.pp.normalize_per_cell(adata, counts_per_cell_after=1e4)
sc.pp.log1p(adata)

# Scale
# sc.pp.scale(adata, max_value=10)
```
